```python
import os

from bagpy import bagreader
```

```python
main_dir = '../data/pb_rosbag/'
surface_dir_paths = os.listdir(main_dir)
for surface_dir_path in surface_dir_paths:
    for trajectory in ('/kwadrat/', '/osemka/'):
        bag_paths = [file for file in os.listdir(main_dir + surface_dir_path + trajectory) if file.endswith(".bag")]
        for bag_path in bag_paths:
            b = bagreader(main_dir + surface_dir_path + trajectory + bag_path)
            for topic in b.topics:
                b.message_by_topic(topic)
```
