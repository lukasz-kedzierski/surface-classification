```python
import numpy as np                 # v 1.19.2
import matplotlib.pyplot as plt    # v 3.3.2

# Enter x and y coordinates of points and colors
xs = [0.2, -0.2]
ys = [0, 0]
colors = ['k', 'k']

# Select length of axes and the space between tick labels
xmin, xmax, ymin, ymax = -0.5, 0.5, -0.3, 0.3
ticks_frequency = 0.2

# Plot points
fig, ax = plt.subplots(figsize=(10, 10))
# ax.scatter(xs, ys, c=colors, s=50000, facecolors='none', edgecolors='r')
circle1 = plt.Circle((-0.2, 0), 0.2, color='k', fill=False)
circle2 = plt.Circle((0.2, 0), 0.2, color='k', fill=False)
ax.add_patch(circle1)
ax.add_patch(circle2)

# Set identical scales for both axes
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax), aspect='equal')

# Set bottom and left spines as x and y axes of coordinate system
ax.spines['bottom'].set_position(('outward', 0))
ax.spines['left'].set_position(('outward',0))

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('y coordinate [m]')
ax.set_ylabel('x coordinate [m]')

# Create custom major ticks to determine position of tick labels
x_ticks = np.arange(xmin+0.1, xmax, ticks_frequency)
y_ticks = np.arange(ymin+0.1, ymax, ticks_frequency)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

# Create minor ticks placed at each integer to enable drawing of minor grid
# lines: note that this has no effect in this example with ticks_frequency=1
minor_x_ticks = np.arange(xmin, xmax, 0.1)
minor_y_ticks = np.arange(ymin, ymax, 0.1)
ax.set_xticks(minor_x_ticks, minor=True)
ax.set_yticks(minor_y_ticks, minor=True)

# Draw major and minor grid lines
ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

# Draw arrows
arrow_fmt = dict(markersize=4, color='black', clip_on=False)
ax.plot((1), (-0.3), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
ax.plot((-0.5), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

plt.show()
```
